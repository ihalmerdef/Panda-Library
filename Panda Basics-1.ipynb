```python
#1. Importing the required libraries.
import pandas as pd
import numpy as np
import csv
```

```python
#2. Getting the data-set from a different source (my desktop) and displaying the dataset.
df=pd.read_excel("/Users/SCHOOL/Desktop/patientslikeme/Test.xlsx")
```

```python
df
```

```python
#3. Removing the unused or irrelevant columns.
#df = df.drop(['Random Values'], axis=1, errors='ignore')
d= df.dropna(axis=1, how='any', thresh=None, subset=None, inplace=False)
# dropping null value columns to avoid errors 
#data.dropna(inplace = True) 
```

```python
d
```

```python
d['Added']=[0,1,2]
```

```python
d
```

```python
#4. Renaming the column names as per our convenience.
d.rename(columns={"Added": "added"})
```

```python
#5. Replacing the value of the rows and make it more meaningful.//done
d.replace(to_replace='Computer Science', value='CS Department', regex=True)
```

```python
d
```

```python
#6. Spliting rows//
new = d["Name"].str.split(" ", n = 1, expand = True) 
new
# making separate first name column from new data frame 

d["First Name"]= new[0] 
d["Sedond Name"]= new[1] 

# Dropping old Name columns 
d.drop(columns =["Name"], inplace = True)
d
```

```python
#Reorder columns
d = d[["First Name", "Sedond Name", "University","College","Department","ID","Added"]]
d
```

```python
#Export the table
d.to_excel("output.xlsx")  
```
